```python
# Define age bins and labels  
bins = [18, 30, 40, 50, 60, 70, 80, 100]  
labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']  

# Cut the age column into bins  
cust['age_group'] = pd.cut(cust['age'], bins=bins, right=False, labels=labels)  

# Count occurrences in each age group  
age_group_counts = cust['age_group'].value_counts().sort_index()  

# Create a bar chart using Plotly  
fig2 = px.bar(age_group_counts, x=age_group_counts.index, y=age_group_counts.values,  
              labels={'index': '<b>Age Groups', 'y': '<b>Number of Customers'},  
              title='Age Distribution of<b><br> Customers',
              template='seaborn',
              text_auto=True)  

fig2.show()
```

```python
# Define income bins and labels  
bins = [0, 40000, 50000, 60000, 70000, 80000, 90000, 100000, float('inf')]  
labels = ['< 40,000', '40,000 - 49,999', '50,000 - 59,999', '60,000 - 69,999',  
          '70,000 - 79,999', '80,000 - 89,999', '90,000 - 99,999', '100,000+']  

# Cut the income column into bins  
cust['income_group'] = pd.cut(cust['income'], bins=bins, right=False, labels=labels)  

# Count occurrences in each income group  
income_group_counts = cust['income_group'].value_counts().sort_index()  

# Create a bar chart using Plotly  
fig3 = px.bar(income_group_counts, x=income_group_counts.index, y=income_group_counts.values,  
              labels={'index': '<b>Income Groups', 'y': '<b>Number of Customers'},  
              title='Income Distribution of<b><br> Customers',
              template='seaborn',
              text_auto=True)

fig3.show()
```

<h3>Tenure Segmentation

```python
cust['became_member_on']= pd.to_datetime(cust['became_member_on'],format= '%Y%m%d')
```

```python
#Creating a column of days and years in duration
from datetime import datetime
today = pd.to_datetime(datetime.now().date()) #getting today's date

cust['Membership duration in days'] = (today - cust['became_member_on']).dt.days
cust['Membership duration in years'] = cust['Membership duration in days']/365
```

<h3>Offer Completion Rate:<br><h5>The % of customers who responded to an offer

```python
off.head()
```

```python
event['offer id'] = event.loc[event['value'].str.contains('offer id'),'value'].str.split(expand=True)[2]
```

```python
event['amount'] = event.loc[event['value'].str.contains('amount'),'value'].str.split(expand=True)[1]
```

```python
event['offer id'] = event['offer id'].str.replace('}','').str.replace("'","", regex=False)
```

```python
event['amount'] = event['amount'].str.replace('}','').astype('float')
```

```python
print('total event records',len(event))
print('total customer records',len(cust))
```

```python
#join the customer_id with the events table include all the customers from the even table
combo1 = event.merge(cust,how='left',left_on='customer_id',right_on='customer_id')
```

```python
#Merging the combined table of customers and events with the offer tables to get the final table for analysis
final = combo1.merge(off, how='left',left_on='offer id', right_on='offer_id')
```

```python
# Function to determine Response Rate  
def determine_response_rate(hours):  
    if hours < 24:  
        return 'Immediate response'  
    elif 24 <= hours < 120:  
        return 'Early response'  
    else:  
        return 'Delayed response'  

# Creating the 'Response Rate' column  
final['Response Rate'] = final['time'].apply(determine_response_rate)  
 
```

```python
final.groupby(['age_group','Response Rate'],as_index=False)['customer_id'].count()
```

```python
px.bar(final.groupby(['age_group','Response Rate'],as_index=False)['customer_id'].count().sort_values('age_group',ascending=False),
       x='customer_id',y='age_group',color='Response Rate', template='seaborn',
       text_auto = True,
       labels={'customer_id':'Count','age_group':'Age Group'},
       title='Which Age Group of the customers had better by<b><br>Response Rate')
```

```python
import dash
import dash_bootstrap_components as dbc
from dash import html, dcc
import plotly.express as px
from dash.dependencies import Input, Output
```

```python
colors = ['gold', 'mediumturquoise']
```

<h2>Dash app code

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.CERULEAN])

Sidebar_style = {            #this is styling features stored inside list to format the navigation section
    'position':'fixed',
    'top':0,
    'left':0,
    'bottom':0,
    'width': '16rem', #this is showing the width of the side navigation bars that we have created
    'padding':'2rem 1rem',
    'background-color':'#f8f9fa'
}

content_style = {   #this is the styling features for the content that will come in the centre of the page
    'margin-left':'18rem', #this distance shows the centre conten will have with the side navigation bar
    'margin-right':'2rem',
    'padding':'2rem 1rem'
}

#The below sidebar visuals can be easily achieved using the dbc.Card method which is much easier(also shown in the tutorial)

sidebar = html.Div( 
    [
        html.H2('Analysis', className='display-4 text-dark'), #this gives a heading with different classname to edit text
        html.Hr(), #this creates a horizontal line
        html.P(
            'Data summary based on the following segmentation', className='lead text-dark'
        ),
        
        dbc.Nav(
            [
               dbc.NavLink('Customer Details', href='/customer-details', active='exact', className='text-dark'),
               dbc.NavLink('Offer Details', href='/offer-details', active='exact', className='text-dark')
            ],
            vertical=True,
            pills=True  #give a backkground color to the navigation sectionn on True
        ),
    ],
    style= Sidebar_style , #this styling is of the Div component
)

content = html.Div(id='page-content', children=[],style= content_style) #leaving children empty as the content goes inside it

app.layout = html.Div([dcc.Location(id='url'),sidebar,content])

@app.callback(
    Output('page-content','children'),
    Input('url','pathname')
)

def render_page_content(pathname):
    if pathname == '/customer-details':
        return[
            html.H1('Maven Cafe Rewards Summary',
                    style= {'textAlign':'center'}),
            html.H3('Customer Details',style={'textAlign':'center'}),
            html.Br(),
            html.Div([
                html.H4('Total Customers | 16994', className='col-4'),
                html.H4('Total Type of Offers | 10',className='col-4')],className='row'),
            html.Div([
            dcc.Graph(id='piechart',className='col-6',
                    
                        figure = px.pie((cust['gender'].value_counts(normalize=True)*100).reset_index(),
                        values='gender',
                        names='index',
                        hole=0.4,
                        title='Customers Distribution <br><b>By Gender').update_traces(textposition = 'inside',
                                                                                       textfont_size=17,
                                                                                        marker=dict(colors=colors,
                                                                                                    line=dict(color='#000000', width=2)))
                        
                        # figure.update_traces(textposition = 'inside',textfont_size=17,
                        #         marker=dict(colors=colors, line=dict(color='#000000', width=2)))
                        ),
            dcc.Graph(id='bargraph', className='col-6',
                     figure= px.bar(age_group_counts, x=age_group_counts.index, y=age_group_counts.values,  
                            labels={'index': '<b>Age Groups', 'y': '<b>Number of Customers'},  
                            title='Age Distribution of<b><br> Customers',
                            template='seaborn',
                            text_auto=True))
            ],className='row'),
            
            dcc.Graph(id='bargraph2',
                      figure= px.bar(income_group_counts, x=income_group_counts.index, y=income_group_counts.values,  
                        labels={'index': '<b>Income Groups', 'y': '<b>Number of Customers'},  
                        title='Income Distribution of<b><br> Customers',
                        template='seaborn',
                        text_auto=True)),
            dcc.Graph(id='histogram',
                      figure=px.histogram(cust,x='Membership duration in years',color='gender',
                    nbins=50,barmode='overlay',
                    template='seaborn',
                    title='Membership distribution of customers by <br><b>Gender').update_traces(marker_line_width=1,marker_line_color="white"))               
        ]
        
    elif pathname == '/offer-details':
        return[
            html.H1('Maven Cafe Rewards Summary',
                    style= {'textAlign':'center'}),
            html.H3('Offer Details',style={'textAlign':'center'}),
            html.Br(),
            html.Div([html.H6('Total Revenue | $177.5K',className='col-2'),
                    html.H6('Total Transactions | 139K',className='col-2'),
                    html.H6('Total Viewed | 57.7K',className='col-2'),
                    html.H6('Total Received | 76.2K',className='col-2'),
                    html.H6('Total Completed | 33.5K',className='col-2')]
                     , className='row'),
            html.Div([
            dcc.Graph(id='bargraph3', className='col-4',
                      figure=px.bar((final['offer_type'].value_counts(normalize=True)*100).reset_index(),
                        x= 'index',y='offer_type',
                        template='seaborn',
                        labels={'index':'<b>Offer Type','offer_type':'<b>Conversion rate %'},
                        title='Conversion Rate by<b><br>Offer Type',
                        text_auto= True).update_traces(texttemplate = '<b>%{value:.0f}%')),
            
            dcc.Graph(id='bargraph5',className='col-8',
                      figure=px.bar((final['channels'].value_counts(normalize=True)*100).reset_index().sort_index(ascending=False),
                        x= 'channels',y='index',
                        template='seaborn',
                        labels={'index':'<b>channels','channels':'<b>Conversion rate %'},
                        title='Conversion Rate by<b><br>Channels',
                        text_auto= True).update_traces(texttemplate = '<b>%{value:.0f}%'))
        ],className='row'),
            
            
            dcc.Graph(id='bargraph4',
                     figure= px.bar(final.groupby(['age_group','Response Rate'],as_index=False)['customer_id'].count().sort_values('age_group',ascending=False),
                    x='customer_id',y='age_group',color='Response Rate', template='seaborn',
                    text_auto = True,
                    labels={'customer_id':'<b>Count','age_group':'<b>Age Group'},
                    title='Which Age Group of the customers had better by<b><br>Response Rate'))            
        ]
        

if __name__== '__main__':
    app.run_server(debug=True, port=3040)
    
# port name: http://127.0.0.1:3040/
```
